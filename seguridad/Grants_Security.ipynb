Widgets - Environment

```python
%python
dbutils.widgets.removeAll()
```

```python
%sql
CREATE widget TEXT storageName DEFAULT "adlsinhueproject";
CREATE widget TEXT catalogo DEFAULT "catalog_project_ssm";
CREATE widget TEXT container DEFAULT "raw";
CREATE widget TEXT external_locations_raw DEFAULT "exlt-raw";
CREATE widget TEXT external_locations_bronze DEFAULT "exlt-bronze";
CREATE widget TEXT esquema_bronze DEFAULT "bronze";
CREATE widget TEXT esquema_silver DEFAULT "silver";
CREATE widget TEXT esquema_golden DEFAULT "golden";

CREATE widget TEXT table_silver DEFAULT "renovacion_prestamo_transformed";
CREATE widget TEXT table_golden DEFAULT "golden_renovacion_prestamo";

CREATE widget TEXT group_data_analysts DEFAULT "data_analysts";
CREATE widget TEXT group_data_engineers DEFAULT "data_engineers";
```

```python
%python
storageName = dbutils.widgets.get("storageName")
catalogo = dbutils.widgets.get("catalogo")
container = dbutils.widgets.get("container")
external_locations_raw = dbutils.widgets.get("external_locations_raw")
external_locations_bronze = dbutils.widgets.get("external_locations_bronze")
esquema_bronze = dbutils.widgets.get("esquema_bronze")
esquema_silver = dbutils.widgets.get("esquema_silver")
esquema_golden = dbutils.widgets.get("esquema_golden")

table_silver = dbutils.widgets.get("table_silver")
table_golden = dbutils.widgets.get("table_golden")

group_data_analysts = dbutils.widgets.get("group_data_analysts")
group_data_engineers = dbutils.widgets.get("group_data_engineers")
```

```python
spark.sql(f"REVOKE SELECT ON SCHEMA {catalogo}.{esquema_bronze} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE ON SCHEMA {catalogo}.{esquema_bronze} FROM {group_data_engineers}")
spark.sql(f"REVOKE USE SCHEMA ON SCHEMA {catalogo}.{esquema_bronze} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE TABLE ON SCHEMA {catalogo}.{esquema_bronze} FROM {group_data_engineers}")

spark.sql(f"REVOKE SELECT ON SCHEMA {catalogo}.{esquema_silver} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE ON SCHEMA {catalogo}.{esquema_silver} FROM {group_data_engineers}")
spark.sql(f"REVOKE USE SCHEMA ON SCHEMA {catalogo}.{esquema_silver} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE TABLE ON SCHEMA {catalogo}.{esquema_silver} FROM {group_data_engineers}")

spark.sql(f"REVOKE SELECT ON SCHEMA {catalogo}.{esquema_golden} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE ON SCHEMA {catalogo}.{esquema_golden} FROM {group_data_engineers}")
spark.sql(f"REVOKE USE SCHEMA ON SCHEMA {catalogo}.{esquema_golden} FROM {group_data_engineers}")
spark.sql(f"REVOKE CREATE TABLE ON SCHEMA {catalogo}.{esquema_silver} FROM {group_data_engineers}")

#External Locations
spark.sql(f"REVOKE READ FILES ON EXTERNAL LOCATION `{external_locations_raw}` FROM {group_data_engineers}")
spark.sql(f"REVOKE WRITE FILES ON EXTERNAL LOCATION `{external_locations_raw}` FROM {group_data_engineers}")

spark.sql(f"REVOKE READ FILES ON EXTERNAL LOCATION `{external_locations_bronze}` FROM {group_data_engineers}")
spark.sql(f"REVOKE WRITE FILES ON EXTERNAL LOCATION `{external_locations_bronze}` FROM {group_data_engineers}")
```

```python
#Grupo Data Analyst
spark.sql(f"REVOKE SELECT ON SCHEMA {catalogo}.{esquema_bronze} FROM {group_data_analysts}")

spark.sql(f"REVOKE SELECT ON TABLE {catalogo}.{esquema_silver}.{table_silver} FROM {group_data_analysts}")

spark.sql(f"REVOKE SELECT ON TABLE {catalogo}.{esquema_golden}.{table_golden} FROM {group_data_analysts}")

spark.sql(f"REVOKE READ FILES ON EXTERNAL LOCATION `{external_locations_bronze}` FROM {group_data_analysts}")
```
