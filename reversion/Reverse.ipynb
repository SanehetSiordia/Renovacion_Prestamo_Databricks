```python
%python
dbutils.widgets.removeAll()
```

```python
%sql
CREATE widget TEXT storageName DEFAULT "adlsinhueproject";
CREATE widget TEXT catalogo DEFAULT "catalog_project_ssm";
CREATE widget TEXT esquema_bronze DEFAULT "bronze";
CREATE widget TEXT esquema_silver DEFAULT "silver";
CREATE widget TEXT esquema_golden DEFAULT "golden";

CREATE widget TEXT table_bronze_a DEFAULT "renovacion_prestamo_clientes_info_operacional";
CREATE widget TEXT table_bronze_b DEFAULT "renovacion_prestamo_clientes_score_financiero";
CREATE widget TEXT table_silver DEFAULT "renovacion_prestamo_transformed";
CREATE widget TEXT table_golden DEFAULT "golden_renovacion_prestamo";
```

```python
%python
storageName = dbutils.widgets.get("storageName")
catalogo = dbutils.widgets.get("catalogo")
esquema_bronze = dbutils.widgets.get("esquema_bronze")
esquema_silver = dbutils.widgets.get("esquema_silver")
esquema_golden = dbutils.widgets.get("esquema_golden")

table_bronze_a = dbutils.widgets.get("table_bronze_a")
table_bronze_b = dbutils.widgets.get("table_bronze_b")
table_silver = dbutils.widgets.get("table_silver")
table_golden = dbutils.widgets.get("table_golden")
```

```python
%python

ruta_bronze = f"abfss://{esquema_bronze}@{storageName}.dfs.core.windows.net"
ruta_silver = f"abfss://{esquema_silver}@{storageName}.dfs.core.windows.net"
ruta_gold = f"abfss://{esquema_golden}@{storageName}.dfs.core.windows.net"
```

Eliminacion tablas

```python
#DROP TABLES
spark.sql(f"DROP TABLE IF EXISTS {catalogo}.{esquema_bronze}.{table_bronze_a}")
spark.sql(f"DROP TABLE IF EXISTS {catalogo}.{esquema_bronze}.{table_bronze_b}")
spark.sql(f"DROP TABLE IF EXISTS {catalogo}.{esquema_silver}.{table_silver}")
spark.sql(f"DROP TABLE IF EXISTS {catalogo}.{esquema_golden}.{table_golden}")
```

```python
%python
## REMOVE DATA TABLE
dbutils.fs.rm(f"{ruta_bronze}/{table_bronze_a}", True)
dbutils.fs.rm(f"{ruta_bronze}/{table_bronze_b}", True)
dbutils.fs.rm(f"{ruta_silver}/{table_silver}", True)
dbutils.fs.rm(f"{ruta_gold}/{table_golden}", True)
```

Eliminacion de Catalogo

```python
#DROP SCHEMA
spark.sql(f"DROP SCHEMA IF EXISTS {catalogo}.{esquema_bronze} CASCADE")
spark.sql(f"DROP SCHEMA IF EXISTS {catalogo}.{esquema_silver} CASCADE")
spark.sql(f"DROP SCHEMA IF EXISTS {catalogo}.{esquema_golden} CASCADE")
```

```python
#DROP CATALOG
spark.sql(f"DROP CATALOG IF EXISTS {catalogo} CASCADE")
```
